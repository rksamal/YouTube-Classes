### 7. Data Manipulation

```python
import pandas as pd

# Sample DataFrame 1
df1 = pd.DataFrame({
    'common_column': [1, 2, 3, 4],
    'column_df1': ['A', 'B', 'C', 'D'],
    'value_df1': [10, 20, 30, 40]
})

# Sample DataFrame 2
df2 = pd.DataFrame({
    'common_column': [2, 3, 4, 5],
    'column_df2': ['X', 'Y', 'Z', 'W'],
    'value_df2': [50, 60, 70, 80]
})

# Merging DataFrames based on a common column
merged_df = pd.merge(df1, df2, on='common_column')

# Displaying the result
print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)
print("\nMerged DataFrame:")
print(merged_df)
```

```python
import pandas as pd

# Sample DataFrame 1
df1 = pd.DataFrame({
    'common_column': [1, 2, 3, 4],
    'column_df1': ['A', 'B', 'C', 'D'],
    'value_df1': [10, 20, 30, 40]
})

# Sample DataFrame 2
df2 = pd.DataFrame({
    'common_column': [5, 6, 7, 8],
    'column_df2': ['X', 'Y', 'Z', 'W'],
    'value_df2': [50, 60, 70, 80]
})

# Concatenating DataFrames along rows
concatenated_df = pd.concat([df1, df2], axis=0)

# Displaying the result
print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)
print("\nConcatenated DataFrame:")
print(concatenated_df)
```
