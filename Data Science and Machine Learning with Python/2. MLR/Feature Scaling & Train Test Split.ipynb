# Feature Scaling

```python
#Importing the libraries
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv('50_startups.csv')
```

```python
dataset.head()
```

```python
dataset_numerical = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']]
```

```python
dataset_numerical.head()
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
df_scaled = pd.DataFrame(standard_scaler.fit_transform(dataset_numerical), columns=dataset_numerical.columns)
df_scaled
```

```python
from sklearn.preprocessing import MinMaxScaler
norm_scaler = MinMaxScaler()
df_norm = pd.DataFrame(norm_scaler.fit_transform(dataset_numerical), columns=dataset_numerical.columns)
df_norm
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

#Setting the value for X and Y
X = df_norm.drop(['Profit'], axis=1)
y = df_norm['Profit']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```
